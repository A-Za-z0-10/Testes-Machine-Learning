```python
from sklearn.preprocessing import Imputer# Imputer vai colocar um valor de entrada responsavel pelos valores faltantes
imputer = Imputer(missing_values ='NaN', strategy = 'mean', axis =0)
imputer = imputer.fit(previsores[:, 0:3])# metodo fit encaixa a nova variavel na base de dados
previsores[:, 0:3] = imputer.transform(previsores[:, 0:3])
```

```python
# TIPOS DE ESCALONAMENTO

# PADRONIZAÇÃO - STANDARDISATION 

#X = X - MEDIA(X) / DESVIO PADRAO(X)

# NORMALIZAÇÃO - NORMALIZATION

#X = X - MINIMO(X) / MAXIMO(X) - MINIMO(X)


#DIGAMOS QUE EU TENHA UMA PESSOA DE 20 ANOS O MENOR VALOR DE IDADE 10 ANOS
#A PESSOA DE MAIOR IDADE 60 UTILIZANDO A NORMALIZACAO O VALOR DA MUDANCA DE ESCALA SERA 0.2

# 20 - 10 / 60 - 10 = 0,2

# distancia euclidiana distancia entre dois pontos
```

```python
# tranformar todos atributos numericos na mesma escala
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```

```python
#divisaõ da base de dados de teste e treinamento
from sklearn.cross_validation import train_test_split

#train_test_split funcao que divisão automatico teste e treinamento
#colocar dentro da funcao train_test_split(base de dados original que queremos fazer a divisão)

#teste_size = quantos % da base de dados vai ser utilizado para teste

#random_state porção da base de dados
```

```python
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25, random_state=0)
```

```python
previsores_treinamento.shape
```

```python
previsores_teste.shape
```

```python
previsores_teste
```

```python
classe_treinamento.shape
```

```python
classe_teste.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classificador = GaussianNB()
classificador.fit(previsores_treinamento, classe_treinamento)
```

```python

#Submeter a tabela previsores_teste a tabela de probabilidade 
#- TEMOS 500 REGISTROS NA TABELA PREVISORES_TESTE
#Esta tabela sera submetida a tabela Naive Bayes vai ser feito as multiplicações e fazer as estimativas
# vai retornar o que foi classificado para cada registro
previsoes = classificador.predict(previsores_teste)
```

```python
previsoes#Resultado da previsão do naive bayes 
```

```python
classe_teste # Resposta correta
```

```python
#verificacao do percentual de acerto do algoritimo 
#COMPARACÃO DA RESPOSTA DO NAIVE BAYES COM A RESPOSTA CORRETA 
from sklearn.metrics import confusion_matrix, accuracy_score
precisao = accuracy_score(classe_teste, previsoes)
```

```python
precisao 
```

```python
#verificar o numero de registro que algoritimo acertou - Matriz confusion
matriz = confusion_matrix(classe_teste, previsoes)


# 428 - 8 linha da classe zero ## não pagou o financiamento #

# o algoritimo considerou 8 registros da classe zero como classe um (428 - 8)

# 23 - 41 linha da classe um ##pagou o financiamento##

#  o algoritimo considerou 23 registros da classe um como classe zero (23- 41)


# dos 500 registros da base de dados teste 469 o algoritmo acertou
```
